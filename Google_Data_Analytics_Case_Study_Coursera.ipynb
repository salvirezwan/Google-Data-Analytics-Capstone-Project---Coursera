```python
# Merging the sleep_data and activity_data based on the 'Id' and the date columns
merged_data = pd.merge(sleep_data[['Id', 'SleepDay', 'TotalMinutesAsleep']],
                       activity_data[['Id', 'ActivityDate', 'Calories']],
                       left_on=['Id', 'SleepDay'],
                       right_on=['Id', 'ActivityDate'],
                       how='inner')

# Checking the first few rows of the merged data to ensure it's correct
print(merged_data.head())

# Creating a scatter plot to analyze the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x=merged_data['TotalMinutesAsleep'], y=merged_data['Calories'], palette='Blues')

# Adding labels and title
plt.xlabel('Total Minutes Asleep')
plt.ylabel('Calories')
plt.title('Relationship Between Total Minutes Asleep and Calories')

# Displaying the plot
plt.show()

# Calculating and displaying the correlation value
correlation = merged_data[['TotalMinutesAsleep', 'Calories']].corr().iloc[0, 1]
print(f"Correlation between Total Minutes Asleep and Calories: {correlation}")

```

```python
# Now, let's proceed with the merging process and analysis again
merged_weight_data = pd.merge(weight_data[['Id', 'Date', 'WeightKg']],
                               activity_data[['Id', 'ActivityDate', 'Calories']],
                               left_on=['Id', 'Date'],
                               right_on=['Id', 'ActivityDate'],
                               how='inner')

# Checking the first few rows of the merged data to ensure it's correct
print(merged_weight_data.head())

# Creating a scatter plot to analyze the relationship (if merged data is not empty)
if not merged_weight_data.empty:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=merged_weight_data['WeightKg'], y=merged_weight_data['Calories'], palette='Blues')

    # Adding labels and title
    plt.xlabel('Weight (kg)')
    plt.ylabel('Calories')
    plt.title('Relationship Between Weight and Calories')

    # Displaying the plot
    plt.show()

    # Calculating and displaying the correlation value
    correlation_weight_calories = merged_weight_data[['WeightKg', 'Calories']].corr().iloc[0, 1]
    print(f"Correlation between Weight and Calories: {correlation_weight_calories}")
else:
    print("No common data to analyze.")
```

```python
# Checking the minimum and maximum values of calories
min_calories = activity_data['Calories'].min()
max_calories = activity_data['Calories'].max()

# Printing the results
print(f"Minimum Calories: {min_calories}")
print(f"Maximum Calories: {max_calories}")
```

```python
# Define calorie ranges with new thresholds
very_active_threshold = 3000  # Threshold for "Very Active"
moderately_active_threshold = 1800  # Threshold for "Moderately Active"

# Categorizing the users based on their calorie consumption
def categorize_activity(calories):
    if calories >= very_active_threshold:
        return 'Very Active'
    elif calories >= moderately_active_threshold:
        return 'Moderately Active'
    else:
        return 'Lightly Active'

# Applying the categorization to the 'Calories' column in activity_data
activity_data['ActivityLevel'] = activity_data['Calories'].apply(categorize_activity)

# Checking the first few rows to verify the new column
print(activity_data[['Calories', 'ActivityLevel']].head())

# Plotting the pie chart
activity_counts = activity_data['ActivityLevel'].value_counts()

plt.figure(figsize=(8, 8))
activity_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#FF9999','#66B3FF','#99FF99'])
plt.title('Distribution of Users Based on Activity Level')
plt.ylabel('')  # Hide the y-axis label
plt.show()
```

```python
# Calculating the average number of steps taken each day in activity_data
average_steps = activity_data['TotalSteps'].mean()

# Displaying the result
print(f"Average total number of steps taken each day: {average_steps:.2f}")
```

```python
# Calculating the average for each relevant column in activity_data including distances
average_sedentary_minutes = activity_data['SedentaryMinutes'].mean()
average_total_distance = activity_data['TotalDistance'].mean()
average_light_active_minutes = activity_data['LightlyActiveMinutes'].mean()
average_fairly_active_minutes = activity_data['FairlyActiveMinutes'].mean()
average_very_active_minutes = activity_data['VeryActiveMinutes'].mean()
average_very_active_distance = activity_data['VeryActiveDistance'].mean()
average_moderately_active_distance = activity_data['ModeratelyActiveDistance'].mean()
average_light_active_distance = activity_data['LightActiveDistance'].mean()

# Displaying the results
print(f"Average total number of sedentary minutes each day: {average_sedentary_minutes:.2f}")
print(f"Average total distance covered each day (in miles): {average_total_distance:.2f}")
print(f"Average number of light active minutes each day: {average_light_active_minutes:.2f}")
print(f"Average number of fairly active minutes each day: {average_fairly_active_minutes:.2f}")
print(f"Average number of very active minutes each day: {average_very_active_minutes:.2f}")
print(f"Average very active distance covered each day (in miles): {average_very_active_distance:.2f}")
print(f"Average moderately active distance covered each day (in miles): {average_moderately_active_distance:.2f}")
print(f"Average light active distance covered each day (in miles): {average_light_active_distance:.2f}")
```

```python
# Counting the number of days of activity data for each user (Id)
user_activity_days = activity_data.groupby('Id')['ActivityDate'].nunique()

# Categorizing users into High, Moderate, and Light User based on the number of days
def categorize_user(days):
    if days >= 21:
        return 'High User'
    elif 11 <= days <= 20:
        return 'Moderate User'
    else:
        return 'Light User'

# Applying the categorization
user_activity_days = user_activity_days.apply(categorize_user)

# Plotting the pie chart for user categories
user_activity_counts = user_activity_days.value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
user_activity_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#FF9999', '#66B3FF', '#99FF99'])
plt.title('Distribution of Users Based on Activity Data Availability')
plt.ylabel('')  # Hide the y-axis label
plt.show()
```
